## Inventory

```python
inv = pd.read_csv('datasets\inventory.csv')
```

```python
inv.head()
```

```python
inv.drop(columns=['RegionID', 'SizeRank', 'RegionType'], inplace=True)
```

```python
inv.columns
```

```python
inv.rename(columns={'RegionName':'region_name', 'StateName':'state_name'}, inplace=True)
```

```python
inv.head()
```

```python
inv.state_name = inv.state_name.fillna('USA')
```

```python
inv.head()
```

```python
inv.info()
```

```python
inv.isna().sum()
```

```python
state_inv = inv.groupby('state_name').mean()
```

```python
state_inv.mean(axis=1)
```

```python
px.scatter(state_inv, log_y=True, title='Mean Inventory of Homes 2018-2022', template='presentation', labels={'state_name': 'State', 'value':'Inventory'})
```

```python
px.scatter(state_inv.loc[:,'1/31/2022':].mean(axis=1), log_y=True, title='Mean Inventory of Homes 2022', template='presentation', labels={'state_name': 'State', 'value':'Inventory'})
```

```python
px.scatter(state_inv.mean(axis=1), log_y=True, title='Overall Mean Inventory 2018-2022')
```

```python
ga_inv = inv.query("state_name=='GA'").drop(columns='state_name')
pa_inv = inv.query("state_name=='PA'").drop(columns='state_name')
md_inv = inv.query("state_name=='MD'").drop(columns='state_name')
```

### Georgia Inventory

```python
ga_inv.head()
```

```python
ga_inv.columns
```

```python
time_frame = ['1/31/2018', '2/28/2018', '3/31/2018', '4/30/2018',
       '5/31/2018', '6/30/2018', '7/31/2018', '8/31/2018', '9/30/2018',
       '10/31/2018', '11/30/2018', '12/31/2018', '1/31/2019', '2/28/2019',
       '3/31/2019', '4/30/2019', '5/31/2019', '6/30/2019', '7/31/2019',
       '8/31/2019', '9/30/2019', '10/31/2019', '11/30/2019', '12/31/2019',
       '1/31/2020', '2/29/2020', '3/31/2020', '4/30/2020', '5/31/2020',
       '6/30/2020', '7/31/2020', '8/31/2020', '9/30/2020', '10/31/2020',
       '11/30/2020', '12/31/2020', '1/31/2021', '2/28/2021', '3/31/2021',
       '4/30/2021', '5/31/2021', '6/30/2021', '7/31/2021', '8/31/2021',
       '9/30/2021', '10/31/2021', '11/30/2021', '12/31/2021', '1/31/2022',
       '2/28/2022', '3/31/2022', '4/30/2022', '5/31/2022', '6/30/2022',
       '7/31/2022', '8/31/2022', '9/30/2022', '10/31/2022', '11/30/2022',
       '12/31/2022', '1/31/2023']
```

```python
px.line(ga_inv, x='region_name', y= time_frame, log_y=True)
```

```python
ga_inv_2018 = ga_inv.loc[:,'1/31/2018': '12/31/2018'].mean(axis=1)
ga_inv_2019 = ga_inv.loc[:,'1/31/2019': '12/31/2019'].mean(axis=1)
ga_inv_2020 = ga_inv.loc[:,'1/31/2020': '12/31/2020'].mean(axis=1)
ga_inv_2021 = ga_inv.loc[:,'1/31/2021': '12/31/2021'].mean(axis=1)
ga_inv_2022 = ga_inv.loc[:,'1/31/2022': '12/31/2022'].mean(axis=1)
```

```python
ga_inv_2018.head()
```

```python
ga_inv_2018_full = pd.concat([ga_inv.region_name, ga_inv_2018], axis=1)
ga_inv_2019_full = pd.concat([ga_inv.region_name, ga_inv_2019], axis=1)
ga_inv_2020_full = pd.concat([ga_inv.region_name, ga_inv_2020], axis=1)
ga_inv_2021_full = pd.concat([ga_inv.region_name, ga_inv_2021], axis=1)
ga_inv_2022_full = pd.concat([ga_inv.region_name, ga_inv_2022], axis=1)
```

```python
ga_inv_2018_full.rename(columns={0:'Mean'}, inplace=True)
ga_inv_2019_full.rename(columns={0:'Mean'}, inplace=True)
ga_inv_2020_full.rename(columns={0:'Mean'}, inplace=True)
ga_inv_2021_full.rename(columns={0:'Mean'}, inplace=True)
ga_inv_2022_full.rename(columns={0:'Mean'}, inplace=True)
```

```python
merge1 = pd.concat([ga_inv_2018_full, ga_inv_2019], axis=1)
merge2 =pd.concat([merge1, ga_inv_2020], axis=1)
merge3 = pd.concat([merge2, ga_inv_2021], axis=1)
final_merge = pd.concat([merge3, ga_inv_2022], axis=1)
```

```python
final_merge.columns=['region_name','2018', '2019', '2020', '2021', '2022']
```

```python
px.scatter(final_merge, x='region_name', y=['2018', '2019', '2020', '2021', '2022'], log_y=True, title='Georgia Mean Home Inventory', template='plotly')
```

```python
px.box(final_merge, x='region_name', y=['2018', '2019', '2020', '2021', '2022'], log_y=True, title='Georgia Mean Home Inventory', template='plotly')
```
